## More rules

```python
# generate normal distribution
from scipy.stats import norm

breaks = 2

np.random.seed(0)
X = np.random.normal(0, 1, (10_000, 2))


mean = np.nanmean(X, axis=0)
std = np.nanstd(X, axis=0)
brks = norm.ppf(np.linspace(0, 1, breaks+2))[1:-1]
print(brks)
num_features = len(mean)

# if generated_columns is None:
#     generated_columns = np.repeat(True, num_features)
# assert generated_columns.shape == (num_features,), "generated_columns has wrong shape {}".format(generated_columns.shape)
column_names = None
if column_names is None:
    column_names = ["X[%d]" % i for i in range(num_features)]

ranges_all = []
new_ranges_all = []
for i in range(num_features):
    ranges = []
    if False:
        pass
    # if not generated_columns[i]:
    #     continue
    # if is_categorical(X[:, i]):
    #     categories = np.unique(X[:, i][~np.isnan(X[:, i])])
    #     for cat in categories:
    #         rule = DSRule(lambda x, i=i, k=cat: x[i] == k, f"{column_names[i]} = {cat}")
    #         self.add_rule(rule, method=self.maf_method)
    else:
        # First rule
        v = mean[i] + std[i] * brks[0]
        # rule = DSRule(lambda x, i=i, v=v: x[i] <= v, f"{column_names[i]} < {v:.3f}")
        # self.add_rule(rule, method=self.maf_method)
        # Mid rules
        ranges = [v]
        for j in range(1, len(brks)):
            vl = v
            v = mean[i] + std[i] * brks[j]
            ranges.append(vl)
            # rule = DSRule(lambda x, i=i, vl=vl, v=v: vl <=
                            # x[i] < v, f"{vl:.3f} < {column_names[i]} < {v:.3f}")
            # self.add_rule(rule, method=self.maf_method)
        ranges.append(v)
        
        ranges = list(set(ranges))
        ranges_all.append(ranges)
        # Last rule
        # rule = DSRule(lambda x, i=i, v=v: x[i] > v, f"{column_names[i]} > {v:.3f}")
        # self.add_rule(rule, method=self.maf_method)
        
        # new ranges (in between)
        if breaks != 2:
            raise NotImplementedError("Only 2 breaks are supported for now")


        # # first range 16.7 percent to 50 percent
        # rule = DSRule(lambda x, i=i, vl=mean[i] + 1.5 * std[i] * brks[0], v=mean[i]: vl <=
        #                     x[i] < v, f"{vl:.3f} < {column_names[i]} < {v:.3f}")
        # self.add_rule(rule, method=self.maf_method)
        
        # # second range 50 percent to 83.3 percent
        # rule = DSRule(lambda x, i=i, vl=mean[i], v=mean[i] + 1.5 * std[i] * brks[-1]: vl <=
        #                     x[i] < v, f"{vl:.3f} < {column_names[i]} < {v:.3f}")

        multiplier = (100 / 3) / (50 - 100/3)
        # print(multiplier)
        new_ranges = [mean[i] + multiplier * std[i] * brks[0], mean[i], mean[i] + multiplier * std[i] * brks[-1]]
        new_ranges_all.append(new_ranges)

# print(ranges_all)
# print(new_ranges_all)

fig = px.histogram(X[:, 0])


# add vertical lines in ranges
for r in ranges_all[0]:
    fig.add_vline(x=r, line_dash="dash", line_color="red")

# # add vertical lines in ranges
for r in new_ranges_all[0]:
    fig.add_vline(x=r, line_dash="dash", line_color="orange")

to_highlight_left = new_ranges_all[0][0]
to_highlight_right = new_ranges_all[0][-1]

# plotly highlight

fig.add_vrect(
    x0=to_highlight_left, x1=to_highlight_right,
    fillcolor="LightSalmon", opacity=0.5, line_width=0
)

x0 = X[:, 0]





print(((x0 <= mean[0]) & (x0 >= new_ranges_all[0][0])).sum() / len(x0))
print(((x0 >= mean[0]) & (x0 <= new_ranges_all[0][-1])).sum() / len(x0))

fig.show()
```

## Quick report

```python
folder = "results_3_breaks"

folders = os.listdir(folder)


res_with = {"roc_auc": [], "avg_precision": [], "rules": [], "dataset": [], "f1": [], "accuracy": [], "in_between_rules": []}
res_without = {"roc_auc": [], "avg_precision": [], "rules": [], "dataset": [], "f1": [], "accuracy": [], "in_between_rules": []}

for f in folders:
    files = os.listdir(os.path.join(folder, f))
    
    for file in files:
        if "True" in file:
            # print(os.path.join(folder, f, file))
            
            
            with open(os.path.join(folder, f, file), "r") as _f:
                data_with = json.load(_f)
                data_with = data_with["dst_results"]
                res_with["f1"].append(data_with["test_f1"])
                res_with["accuracy"].append(data_with["test_accuracy"])
                res_with["roc_auc"].append(data_with["test_roc_auc"])
                res_with["avg_precision"].append(data_with["test_avg_precision"])
                res_with["rules"].append(data_with["rules"])
                res_with["dataset"].append(f)
                res_with["in_between_rules"].append(True)   
        else:
            with open(os.path.join(folder, f, file), "r") as _f:
                data_without = json.load(_f)
                data_without = data_without["dst_results"]
                res_without["f1"].append(data_without["test_f1"])
                res_without["accuracy"].append(data_without["test_accuracy"])
                res_without["roc_auc"].append(data_without["test_roc_auc"])
                res_without["avg_precision"].append(data_without["test_avg_precision"])
                res_without["rules"].append(data_without["rules"])
                res_without["dataset"].append(f)
                res_without["in_between_rules"].append(False)
        
    
                
```

```python
with_df = pd.DataFrame(res_with)
without_df = pd.DataFrame(res_without)

# combine both
df = pd.concat([with_df, without_df])


# df = df[~df["dataset"].isin(["adult", "bank-full"])]

fig = px.bar(df, x="dataset", y="roc_auc", color="in_between_rules", barmode="group")

# zoom to y axis to see the difference
fig.update_yaxes(range=[0.8, 1])

# title
fig.update_layout(title="ROC AUC for datasets with and without in between rules (y axis zoomed in)")


fig = px.bar(df, x="dataset", y="avg_precision", color="in_between_rules", barmode="group")

# zoom to y axis to see the difference
fig.update_yaxes(range=[0.7, 1])

# title
fig.update_layout(title="Avg. Precision for datasets with and without in between rules (y axis zoomed in)")
```
